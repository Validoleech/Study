# Задача 1

1. Реализовать класс CustomList наследованием от list

При этом:

- CustomList должен наследоваться от встроенного списка list;
- экземпляры CustomList можно складывать друг с другом и с обычными списками:
```
CustomList([5, 1, 3, 7]) + CustomList([1, 2, 7])  # CustomList([6, 3, 10, 7])
CustomList([1]) + [2, 5]  # CustomList([3, 5])
[2, 5] + CustomList([1])  # CustomList([3, 5])
```
- экземпляры CustomList поддерживают вычитание между собой и с обычными списками:
```
CustomList([5, 1, 3, 7]) - CustomList([1, 2, 7])  # CustomList([4, -1, -4, 7])
CustomList([1]) - [2, 5]  # CustomList([-1, -5])
[2, 5] - CustomList([1])  # CustomList([1, 5])
```

- результатом сложения/вычитания должен быть новый кастомный список;
- после сложения/вычитания исходные списки не должны изменяться;
- при сложении/вычитании списков разной длины отсутствующие элементы меньшего списка считаются нулями;
- при сравнении (==, !=, >, >=, <, <=) экземмпляров CustomList должна сравниваться сумма элементов списков (сравнение с list не нужно);
- должен быть переопределен str, чтобы выводились элементы списка и их сумма;
- списки можно считать всегда числовыми.

2. Тесты CustomList

````python
#╰( ͡☉ ͜ʖ ͡☉ )つ──☆*:・ﾟ   ฅ^•ﻌ•^ฅ   ʕ•ᴥ•ʔ
class CustomList(list):
    
    def _apply_operation(self, other, operation):
        """
        Применяет поэлементную операцию к CustomList и другому списку, включая классический.

        :param other: Список или CustomList для взаимодействия.
        :param operation: Функция, определяющая операцию (сложение/вычитание).
        :return: Новый кастомный список с результатом операции.
        """
        if not isinstance(other, list):
            return NotImplemented

        # Определяем длину самого длинного списка
        max_len = max(len(self), len(other))

        # Проводим операцию, проверяя, не превосходит ли индекс длину self или other
        result = [
            operation(
                self[i] if i < len(self) else 0, 
                other[i] if i < len(other) else 0)
            for i in range(max_len)
        ]

        return CustomList(result)

    # Сложение
    def __add__(self, other):
        return self._apply_operation(other, lambda x, y: x + y)

    def __radd__(self, other):
        return self + other

    # Вычитание
    def __sub__(self, other):
        return self._apply_operation(other, lambda x, y: x - y)

    def __rsub__(self, other):
        return self._apply_operation(other, lambda x, y: y - x)

    # Равенства
    def __eq__(self, other):
        if isinstance(other, CustomList):
            return sum(self) == self(other)
        return NotImplemented

    def __ne__(self, other):
        return not self == other

    # Сравнения
    def __lt__(self, other):
        if isinstance(other, CustomList):
            return sum(self) < sum(other)
        return NotImplemented

    def __le__(self, other):
        return self < other or self == other

    def __gt__(self, other):
        return not self <= other

    def __ge__(self, other):
        return not self < other

    def __str__(self):
        return f"{super().__str__()} (sum: {sum(self)})" # С помощью super() вызываем __str__ родительского класса list
````

````python
"""customlist_test"""
#╰( ͡☉ ͜ʖ ͡☉ )つ──☆*:・ﾟ   ฅ^•ﻌ•^ฅ   ʕ•ᴥ•ʔ
import pytest

def test_add_custom_list():
    cl1 = CustomList([5, 1, 3, 7])
    cl2 = CustomList([1, 2, 7])
    result = cl1 + cl2
    assert result == CustomList([6, 3, 10, 7])

def test_add_custom_list_with_list():
    cl1 = CustomList([1])
    lst = [2, 5]
    result = cl1 + lst
    assert result == CustomList([3, 5])

def test_add_list_with_custom_list():
    lst = [2, 5]
    cl1 = CustomList([1])
    result = lst + cl1
    assert result == CustomList([3, 5])

def test_sub_custom_list():
    cl1 = CustomList([5, 1, 3, 7])
    cl2 = CustomList([1, 2, 7])
    result = cl1 - cl2
    assert result == CustomList([4, -1, -4, 7])

def test_sub_custom_list_with_list():
    cl1 = CustomList([1])
    lst = [2, 5]
    result = cl1 - lst
    assert result == CustomList([-1, -5])

def test_sub_list_with_custom_list():
    lst = [2, 5]
    cl1 = CustomList([1])
    result = lst - cl1
    assert result == CustomList([1, 5])

def test_equality_custom_list():
    cl1 = CustomList([5, 1, 3, 7])
    cl2 = CustomList([1, 2, 7])
    assert cl1 == cl2  # Сумма одинаковая

def test_inequality_custom_list():
    cl1 = CustomList([5, 1, 3, 7])
    cl2 = CustomList([5, 1, 3])
    assert cl1 != cl2  # Суммы разные

def test_str_custom_list():
    cl = CustomList([1, 2, 3])
    assert str(cl) == "[1, 2, 3] (sum: 6)"
````

# Задача 2

  Подменяться должны атрибуты класса и атрибуты экземпляра класса, в том числе добавленные после выполнения конструктора (dynamic в примере).

```
    class CustomMeta(...):
        pass


    class CustomClass(metaclass=CustomMeta):
        x = 50

        def __init__(self, val=99):
            self.val = val

        def line(self):
            return 100

        def __str__(self):
            return "Custom_by_metaclass"


    assert CustomClass.custom_x == 50
    CustomClass.x  # ошибка

    inst = CustomClass()
    assert inst.custom_x == 50
    assert inst.custom_val == 99
    assert inst.custom_line() == 100
    assert str(inst) == "Custom_by_metaclass"

    inst.x  # ошибка
    inst.val  # ошибка
    inst.line() # ошибка
    inst.yyy  # ошибка

    inst.dynamic = "added later"
    assert inst.custom_dynamic == "added later"
    inst.dynamic  # ошибка
```

2. Тесты метакласса

````python
class CustomMeta(type):
    def __new__(cls, name, bases, dct):
        # Обрабатываем атрибуты класса
        for key, value in list(dct.items()):
            if not key.startswith('__'):  # Исключаем магические методы
                dct[f'custom_{key}'] = value
                del dct[key]  # Удаляем оригинальный атрибут

        # Создаем класс с изменениями
        new_class = super().__new__(cls, name, bases, dct)

        # Обрабатываем атрибуты экземпляра класса
        for key, value in list(new_class.__dict__.items()):
            if not key.startswith('__'):  # Исключаем магические методы
                setattr(new_class, f'custom_{key}', value)
                delattr(new_class, key)  # Удаляем оригинальный атрибут

        return new_class
````

````python
# Создаём класс
class CustomClass(metaclass=CustomMeta):
    x = 50

    def __init__(self, val=99):
        self.val = val

    def line(self):
        return 100

    def __str__(self):
        return "Custom_by_metaclass"
````

````python
# Тесты
import pytest

def test_custom_meta():
    # Проверяем атрибуты класса
    assert hasattr(CustomClass, 'custom_x')  # custom_x существует
    assert CustomClass.custom_x == 50  # Значение custom_x равно 50

    try:
        CustomClass.x  # Доступ к x должен вызвать ошибку
    except AttributeError:
        pass

    # Создаем экземпляр класса
    inst = CustomClass()

    # Проверяем атрибуты экземпляра
    assert hasattr(inst, 'custom_x')  # custom_x существует
    assert inst.custom_x == 50  # Значение custom_x равно 50
    assert hasattr(inst, 'custom_val')  # custom_val существует
    assert inst.custom_val == 99  # Значение custom_val равно 99
    assert inst.custom_line() == 100  # Метод custom_line работает

    # Проверяем строковое представление
    assert str(inst) == "Custom_by_metaclass"

    # Проверка ошибок для атрибутов без префикса
    try:
        inst.x  # Доступ к x должен вызвать ошибку
    except AttributeError:
        pass
    try:
        inst.val  # Доступ к val должен вызвать ошибку
    except AttributeError:
        pass
    try:
        inst.line()  # Метод line() должен вызвать ошибку
    except AttributeError:
        pass
    try:
        inst.yyy  # Доступ к несуществующему атрибуту должен вызвать ошибку
    except AttributeError:
        pass

    # Проверяем динамически добавленные атрибуты
    inst.dynamic = "added later"
    assert inst.custom_dynamic == "added later"
    try:
        inst.dynamic  # Доступ к dynamic должен вызвать ошибку
    except AttributeError:
        pass
````

# Задача 3

1. Нужно сделать три дескриптора для какой-то области интереса (наука, финансы, хобби и тд), но если совсем не получается, то можно использовать шаблона ниже в качестве основы.

```
    class Integer:
        pass

    class String:
        pass

    class PositiveInteger:
        pass

    class Data:
        num = Integer()
        name = String()
        price = PositiveInteger()

        def __init__(...):
            ....
```

2. Тесты метакласса и дескрипторов

````python
# Делаем три дескриптора для катушки индуктивности.

class PositiveValue:
    """Дескриптор для положительных числовых значений."""
    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        return instance.__dict__.get(self.name, None)

    def __set__(self, instance, value):
        if not isinstance(value, (int, float)) or value <= 0:
            raise ValueError(f"{self.name} должно быть положительным числом.")
        instance.__dict__[self.name] = value

class NonNegativeValue:
    """Дескриптор для неотрицательных числовых значений."""
    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        return instance.__dict__.get(self.name, None)

    def __set__(self, instance, value):
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError(f"{self.name} должно быть неотрицательным числом.")
        instance.__dict__[self.name] = value

class StringValue:
    """Дескриптор для строк."""
    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        return instance.__dict__.get(self.name, None)

    def __set__(self, instance, value):
        if not isinstance(value, str):
            raise ValueError(f"{self.name} должно быть строкой.")
        instance.__dict__[self.name] = value


class CoilCoreType:
    """Дескриптор для типа катушки."""
    VALID_TYPES = {"немагнитный", "магнитный"}

    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        return instance.__dict__.get(self.name, None)

    def __set__(self, instance, value):
        if not isinstance(value, str) or value not in self.VALID_TYPES:
            raise ValueError(f"{self.name} должно быть одним из: {', '.join(self.VALID_TYPES)}.")
        instance.__dict__[self.name] = value


class Inductor:
    inductance = PositiveValue()  # Индуктивность
    max_current = PositiveValue()  # Максимальный ток
    coil_type = CoilCoreType()  # Тип катушки
    resistance = NonNegativeValue()  # Сопротивление постоянному току
    quality_factor = NonNegativeValue()  # Коэффициент добротности
    resonance_frequency = PositiveValue()  # Частота резонанса
    temperature_coefficient = NonNegativeValue()  # Температурный коэффициент

    def __init__(self, inductance, max_current, coil_core_type, resistance, quality_factor, resonance_frequency, temperature_coefficient=0):
        self.inductance = inductance
        self.max_current = max_current
        self.coil_core_type = coil_core_type
        self.resistance = resistance
        self.quality_factor = quality_factor
        self.resonance_frequency = resonance_frequency
        self.coil_core_type = coil_core_type
        self.temperature_coefficient = temperature_coefficient

    def __str__(self):
        return (
            f"Тип сердечника: {self.coil_type}, "
            f"Индуктивность: {self.inductance} Гн, "
            f"Макс. ток: {self.max_current} А, "
            f"Rdc: {self.resistance} Ом, "
            f"Q: {self.quality_factor}, "
            f"Резонансная частота: {self.resonance_frequency} Гц, "
            f"Материал сердечника: {self.coil_core_type}, "
            f"Температурный коэффициент: {self.temperature_coefficient}"
        )
````

````python
coil = Inductor(
    inductance=0.01,
    max_current=2,
    resistance=0.05,
    quality_factor=100,
    resonance_frequency=1_000_000,
    coil_core_type="магнитный",
    temperature_coefficient=0.002
)

import pytest

def test_valid_coil():
    coil = Inductor(
        inductance=0.01,
        max_current=2,
        resistance=0.05,
        quality_factor=100,
        resonance_frequency=1_000_000,
        coil_core_type="магнитный",
        temperature_coefficient=0.002
    )
    assert coil.inductance == 0.01
    assert coil.max_current == 2
    assert coil.coil_core_type == "магнитный"
    assert coil.resistance == 0.05
    assert coil.quality_factor == 100
    assert coil.resonance_frequency == 1_000_000
    assert coil.temperature_coefficient == 0.002

def test_invalid_inductance():
    with pytest.raises(ValueError, match="inductance должно быть положительным числом."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)

def test_invalid_max_current():
    with pytest.raises(ValueError, match="max_current должно быть положительным числом."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)

def test_invalid_coil_type():
    with pytest.raises(ValueError, match="coil_core_type должно быть одним из: магнитный, немагнитный."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)

def test_invalid_name():
    with pytest.raises(ValueError, match="name должно быть строкой."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)

def test_invalid_resonance_frequency():
    with pytest.raises(ValueError, match="resonance_frequency должно быть положительным числом."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)

def test_invalid_temperature_coefficient():
    with pytest.raises(ValueError, match="temperature_coefficient должно быть неотрицательным числом."):
        Inductor(inductance=0.01, max_current=2, resistance=0.05, quality_factor=100, resonance_frequency=1_000_000, coil_core_type="магнитный", temperature_coefficient=0.002)
````
